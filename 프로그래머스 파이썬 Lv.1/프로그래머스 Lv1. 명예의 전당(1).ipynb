```python
def solution(k, score):
    answer = []
    score2 = []
    # sort리스트에 점수 넣고 sort 한 후에 상위 3개 뽑아서 그중 최소값 answer에 저장
    
    for i in score : 
        
        score2.append(i)
        score2.sort(reverse=True)
        
        if len(score2) > k:
            
            score2.pop()
            
        answer.append(min(score2))
            
           
            
            
        
        
            
            
            
        
        
            
        

            
    return answer
```

```python
solution(k,score)
```
