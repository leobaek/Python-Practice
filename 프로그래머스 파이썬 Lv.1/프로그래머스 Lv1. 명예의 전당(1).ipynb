```python
def solution(k, score):
    answer = []
    score2 = []
    # sort   sort    3    answer 
    
    for i in score : 
        
        score2.append(i)
        score2.sort(reverse=True)
        
        if len(score2) > k:
            
            score2.pop()
            
        answer.append(min(score2))
            
           
            
            
        
        
            
            
            
        
        
            
        

            
    return answer
```
