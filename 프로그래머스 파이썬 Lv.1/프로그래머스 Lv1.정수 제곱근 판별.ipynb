```python
# 내가 푼 풀이
def solution(n):
    answer = 0
    a=0
#     n ** 0.5로 루트 씌우기
    a = n ** 0.5 
#     a를 정수형으로 바꿨을때 기존 a와 값이 다르면 a+1의 제곱 리턴
#   왜냐하면 소수를 정수형으로 바꾸면 소수점 밑으로는 값이 사라진다.
    if int(a) == a :
        answer = int((a+1)**2)
    else :
        answer = -1
    
    return answer
```

```python
solution(144)
```
