```python
def solution(strings, n):
    answer=[]
    
#     lambda 함수 사용해서 정렬하기
#     sorted 함수의 첫번째 파라미터는 정렬하고 싶은 변수 sorted를 사용했기 때문에 list or tuple
#     두번째 파라미터 key = 의 값 안에는 lambda x(받아오는 단어) : x[n] (받아오는 단어의 인덱스의 값), x는 원본 단어를 나타냄으로 x[n]이 같을경우 다시 원본 문자열로 비교
    answer = sorted(strings,key = lambda x:(x[n],x)) 
        
    return answer
```
