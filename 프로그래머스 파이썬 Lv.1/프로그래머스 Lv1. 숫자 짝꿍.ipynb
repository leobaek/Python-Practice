```python
def solution(X, Y):
    answer = []
    result = []
    
    xy = set(X) & set(Y)
    
    if len(xy) == 0 :
        
        return "-1"
    
    elif len(xy) == 1 and "0" in xy : 
        
        return "0"
    
    for n in xy : 
        
        countx = X.count(n)
        county = Y.count(n)
        
        result.append(min(countx,county)*n)
        
    
    result = sorted(result,reverse=True)
    
    result = "".join(result)
    
    
    
                
                
            
                   
                
    
    
    return result
```

```python
x = "12321"
y = "42531"
```

```python
solution(x,y)
```
