```python
def solution(number, limit, power):
    result=0
    array = []
    # number    2 . ==  
    #   limit  limit .
    #  .
    #   .    x 2     .
    #        -1    .
    
    
    # 1 number 
    for i in range(1,number+1):
        answer = 0
        
        # 1 i  
        for j in range(1,int(i**0.5)+1) :
            
            if (i%j) == 0 :
                answer+=2
                
        if (i ** 0.5)%1 == 0 :
            
            answer -=1
        
        if answer > limit : 
            
            answer = power
            
        result += answer 
            
    return result
```
