```python
def solution(number, limit, power):
    result=0
    array = []
    # number의 제곱근의 약수를 구해서 2를 곱한다. == 무기의 공격력
    # 만약 공격력이 limit보다 높다면 limit으로 설정한다.
    # 모두 더한다.
    # 약수는 쌍으로 이루어진다. 그러므로 제곱근의 약수 x 2 하면 전체 약수의 개수가 나온다.
    # 그러나 완전제곱의 경우에는 중복되는 약수가 하나 나오므로 -1 해줘야 전체약수 개수가 된다.
    
    
    # 1부터 number까지 반복
    for i in range(1,number+1):
        answer = 0
        
        # 1부터 i의 제곱근까지 반복
        for j in range(1,int(i**0.5)+1) :
            
            if (i%j) == 0 :
                answer+=2
                
        if (i ** 0.5)%1 == 0 :
            
            answer -=1
        
        if answer > limit : 
            
            answer = power
            
        result += answer 
            
    return result
```
