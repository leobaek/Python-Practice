```python
def solution(s):
    answer = True
    a=0
    b=0
    
    s = s.lower()
    
    for i in s :
        if i == 'p' :
            a += 1
        elif i=='y' :
            b +=1
        else :
            pass
    if a == b :
        return True
    else :
        return False
    
    
    
```
