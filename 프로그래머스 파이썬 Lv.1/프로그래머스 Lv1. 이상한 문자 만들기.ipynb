```python
def solution(s):
    answer = []
    s = s.split()
    for i in s :
        change_word =''
        for idx, word in enumerate(i) :
            if idx % 2 ==0 :
                change_word+=word.upper()
            else :
                change_word+=word.lower()
        answer.append(change_word)
        
    return " ".join(answer)
        
```
