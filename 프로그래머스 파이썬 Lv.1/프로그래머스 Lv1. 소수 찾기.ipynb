```python
# 에라토스테네스의 체 알고리즘
```

```python
def solution(n):
    answer = 0
    for i in range(2,n+1) :
        if all(i % j != 0 for j in range(2,i)) :
            answer+=1
    return answer
```

```python
solution(10)
```
