```python
def solution(cards1, cards2, goal):
    answer = 'Yes'
    #goal을 탐색하며 cards1[0]과 cards2[0]에 있는지 확인 있다면 goal다음 값을 넣고 확인 마지막에 card1과 card2가 비어있으면 yes
    
    for i in goal :
        if cards1 and i == cards1[0] : 
            cards1.remove(i)
        
        elif cards2 and i == cards2[0]:
            cards2.remove(i)
        
        else : 
            answer = "No"
            break
            
    if cards1 ==[] and cards2 ==[] :
        answer ="Yes"
            
    
    
    
    return answer
```

```python
cards1 = ["i", "water", "drink"]
cards2 = ["want", "to"]
goal = ["i", "want", "to", "drink", "water"]
```

```python
solution(cards1,cards2,goal)
```
