```python
def solution(s):
    
    first_alp = ''
    same = 0
    diff = 0
    
    answer = 0
    
    for alp in s : 
        
        if first_alp == '' :
            
            first_alp = alp
            same +=1
            continue
            
        if first_alp == alp : 
            
            same +=1
        
        else :
            
            diff +=1
            
        if same == diff : 
            
            first_alp = ''
            same = 0
            diff = 0
            answer +=1
    
    
    if first_alp != '' : 
        
        answer +=1
            
    
    return answer
```
