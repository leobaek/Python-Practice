```python
##   

def solution(s, n):
    answer=''
    lower = ('abcdefghijklmnopqrstuvwxyz')
    upper = lower.upper()
    index = 0
    lastindex =0
    for i in s :
        
        if i == " " :
            answer+=" "
            continue
        elif i.isupper() == True:
                index = upper.index(i)
                lastindex = (index+n)%26
                answer += upper[lastindex]
        else:
                index = lower.index(i)
                lastindex = (index+n)%26
                answer += lower[lastindex]
            
        
    
         
    return answer
```
