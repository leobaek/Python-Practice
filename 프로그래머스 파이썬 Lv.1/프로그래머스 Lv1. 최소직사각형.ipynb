```python
def solution(sizes):
    max_width = 0
    max_height = 0
    for i in sizes :
        # width,height   
        width,height = i
        # max_width  max_width,width,height   
        max_width = max(max_width,width,height)
        max_height = max(max_height,min(width,height))
    return max_width*max_height
```

```python
x = [[60, 50], [30, 70], [60, 30], [80, 40]]
```

```python
solution(x)
```
