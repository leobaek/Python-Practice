```python
def solution(sizes):
    max_width = 0
    max_height = 0
    for i in sizes :
        # width,height에 가로와 세로길이 할당
        width,height = i
        # max_width의 값을 max_width값,width값,height값을 비교하여 반복하여 갱신
        max_width = max(max_width,width,height)
        max_height = max(max_height,min(width,height))
    return max_width*max_height
```

```python
x = [[60, 50], [30, 70], [60, 30], [80, 40]]
```

```python
solution(x)
```
