```python
def solution(array, commands):
    answer = []
    for i in commands :
        b = array[i[0]-1:i[1]]
        b = sorted(b)
        b = b[i[2]-1]
        answer.append(b)
        b = []
        
    return answer
```

```python
array =[1,5,2,6,3,7,4]
commands = [[2, 5, 3], [4, 4, 1], [1, 7, 3]]
```

```python
solution(array,commands)
```
