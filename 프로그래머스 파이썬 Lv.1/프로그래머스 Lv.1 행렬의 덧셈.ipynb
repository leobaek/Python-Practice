```python
def solution(arr1, arr2):
    answer = []
    
    for a,b in zip(arr1,arr2):
        sum=[]
        
        for i,j in zip(a,b):
            sum.append(i+j)
            print(sum)
        answer.append(sum)
        
    return answer
```

```python
arr = [[1,2],[2,3]]
arr0 = [[3,4],[5,6]]
```

```python
solution(arr,arr0)
```
