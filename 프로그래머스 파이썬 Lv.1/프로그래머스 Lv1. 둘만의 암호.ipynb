```python
def solution(s, skip, index):
    alp = 'abcdefghijklmnopqrstuvwxyz'
    alp = sorted(set(alp) - set(skip))
    total = dict()
    answer = ''

    for idx, value in enumerate(alp):
        total[idx] = value

    for i in s:
        location = alp.index(i)
        location = (location + index) % len(total)
        answer += total[location]

    return answer
```
