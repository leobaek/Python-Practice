```python
def solution(name, yearning, photo):
    score = []
    answer = dict(zip(name,yearning))
    
    for i in photo :
        result=0
        for j in i :
            result += answer.get(j,0)
            
        score.append(result)
            
            
            
            
            
    
    
    return score
```
