```python
def solution(food):
    answer = ''
    num = ''
    for i in range(len(food)):
        if food[i] >= 2 :
            x = food[i] // 2 
            for j in range(x) :
                num+=str(i)
                
        answer = num + '0' + num[::-1]
                
            
            
            
    
    return answer
```
