```python
def solution(n, arr1, arr2):
    answer = []
    arr1_bin = []
    arr2_bin = []
    row= ''
    for i in arr1 : 
        a = bin(i)[2:]
        a = a.zfill(n)
        arr1_bin.append(a)
    for i in arr2 :
        b = bin(i)[2:]
        b = b.zfill(n)
        arr2_bin.append(b)
    
    for i in range(n) :
        row=''
        for j in range(n) :
            if arr1_bin[i][j] == '1' or arr2_bin[i][j] =='1':
                
                row += '#'
            else :
                row +=" "
        answer.append(row)
            
        
    
    return answer
```
