```python
def solution(dartResult):
    
    
    #     
    result = []
    
    #  
    rule = {
        "S" : 1,
        "D" : 2,
        "T" : 3,
    }
    
    #       
    score = ""
    
    # dartResult 
    for i in dartResult :
        
        #    score 
        if i.isdigit() : 
            
            score += i
        
        # rule    score      rule  
        elif i in rule : 
            
            result.append(int(score) ** rule[i])
            score =""
            
    
            
        # #    -1   
        elif i =="#" :
            
            result[-1] = result[-1] * -1
            
        
        # *     2
        elif i =="*" :
            
            result[-1] = result[-1]*2
            
            
            
            #  result   2         *2
            if len(result) >=2 :
                
                result[-2]= result[-2] * 2
                
    result = sum(result)
        
        
            
            
    
  

        
        
    return result
```

```python
a = '1S*2T*3S'
```

```python
solution(a)
```
