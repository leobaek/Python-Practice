```python
def solution(dartResult):
    
    
    # 각 다트별 결과를 담을 리스트
    result = []
    
    # 제곱해야할 규칙
    rule = {
        "S" : 1,
        "D" : 2,
        "T" : 3,
    }
    
    # 현재 문자가 숫자일 경우 담고 있을 문자열
    score = ""
    
    # dartResult를 순회한다
    for i in dartResult :
        
        # 만약 숫자라면 그대로 score에 더하기
        if i.isdigit() : 
            
            score += i
        
        # rule에 있는 문자라면 현재 score에 있는 문자열을 숫자로 변환 후에 rule에 따라 제곱
        elif i in rule : 
            
            result.append(int(score) ** rule[i])
            score =""
            
    
            
        # #이라면 마지막으로 추가한 데이터에 -1을 곱해 음수로 변환
        elif i =="#" :
            
            result[-1] = result[-1] * -1
            
        
        # *이라면 마지막으로 추가한 데이터에 곱하기 2
        elif i =="*" :
            
            result[-1] = result[-1]*2
            
            
            
            # 마지막으로 result에 추가된 값들이 2개 이상일 경우 마지막을 제외한 가장 최근에 저장된 값에 *2
            if len(result) >=2 :
                
                result[-2]= result[-2] * 2
                
    result = sum(result)
        
        
            
            
    
  

        
        
    return result
```

```python
a = '1S*2T*3S'
```

```python
solution(a)
```
