```python
def solution(nums):
    answer=[]
    result=0

    for i in range(len(nums)) :
        
        for j in range(i+1,len(nums)) :
            
            for k in range(j+1,len(nums)) :
                
                answer.append(nums[i] + nums[j] + nums[k])
                
    
    print(answer)
                
                
    for i in answer :
        
        # j는 2부터 i 직전까지의 수 이고
        # i가 j로 나누어 떨어지지 않는다면 소수이다.
         if all((i%j) != 0 for j in range(2,i)):
                result +=1
                    
                    

    return result
```
