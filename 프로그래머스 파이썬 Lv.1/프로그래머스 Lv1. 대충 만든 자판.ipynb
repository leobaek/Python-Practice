```python
def solution(keymap, targets):
    answer = []
    
    keymap_dict ={}
    
    for i in keymap:
    
        for idx,alp in enumerate(i) :
        
            if (alp in keymap_dict) and (idx+1) > keymap_dict[alp] :
                continue
                
            keymap_dict[alp] = idx+1
            
    for i in targets :
        
        total_click = 0
        
        for alp in i : 
            
            if alp in keymap_dict :
                
                total_click += keymap_dict[alp]
            
            else:
                total_click = -1
                break
                
        answer.append(total_click)
    
                    
        
        
    return answer
```

```python
keymap =["ABACD", "BCEFD"]
targets = ["ABCD","AABB"]
```

```python
solution(keymap,targets)
```
