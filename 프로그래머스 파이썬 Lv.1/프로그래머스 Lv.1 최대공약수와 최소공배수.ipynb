```python
def solution(n, m):
    a=0
    b=0
#    n,m중 작은값부터 0까지 -1간격으로 반복
    for i in range(min(n,m),0,-1):
        if n%i == 0 and m%i==0 :
            a += i
            break
            
#   최소공배수 공식 = 두 수를 곱하고 최대공약수로 나눈다.
    b = (n*m) / a
            
    return [a,b]
```
