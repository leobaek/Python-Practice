```python
def solution(left, right):
    answer = 0
    
    for i in range(left,right+1):
        count=0
        for x in range(1,i+1):
            if i%x ==0:
                count+=1
        if count % 2 ==0 :
                answer+=i
        else:
                answer-=i
            
            
       

            
    return answer
```

```python
solution(13,17)
```
