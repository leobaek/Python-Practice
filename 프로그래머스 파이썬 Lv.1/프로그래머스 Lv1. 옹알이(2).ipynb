```python
def solution(babbling):
   
    says = ["aya", "ye", "woo", "ma"]
    result = 0
   
    for i in range(len(babbling)):
    # babbling    range(len(babbling)) 
   
        for say in says:
            if (say in babbling[i]) and (say*2 not in babbling[i]):
            # babbling[i]  say  say*2    
           
                babbling[i] = babbling[i].replace(say, "*")
                # babbling[i] say "*" 
                # ) "ayamwoo" => "*m*"
       
        if all(char == "*" for char in babbling[i]):
            # babbling[i]  "*"   result += 1
            result += 1
   
    return result
```

```python
a = ["aya", "yee", "u", "maa"]
```

```python
solution(a)
```
