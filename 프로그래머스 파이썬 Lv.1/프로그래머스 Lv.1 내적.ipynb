```python
def solution(a, b):
    answer = 0
    for i in range(0,len(a)) :
        answer+= a[i]*b[i]
        
    
    return answer
```

```python
a=[1,2,3,4]
b=[-3,-1,0,2]
solution(a,b)
```

```python
# 다른 풀이

def solution2(a, b):
    answer=0
    for x,y in zip(a,b) :
        answer+=x*y
        
    return answer
```

```python
solution2(a,b)
```
